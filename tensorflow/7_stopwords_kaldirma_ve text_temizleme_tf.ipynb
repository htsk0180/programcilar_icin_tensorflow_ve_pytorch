```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
sentences = [
 'Ben seni severim ancak senin de beni sevmen gerek.',
 'Kalemimi sana veririm ama sağlam getireceksin.',
 'Sen ve o ikiniz bir takımsınız ama bana karşı oynamayacaksınız.',
 'Benim sâdık yârim kara topraktır'
]
```

```python
tokenizer = Tokenizer(num_words = 100)
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(sentences)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded = pad_sequences(sequences)
```

```python
stopwords = ["ancak", "de", "ama", "ve"]
```

```python
# stopwords'leri cümlelerden çıkartıyoruz
for i in range(len(sentences)):
    for word in stopwords:
        if word in sentences[i]:
            sentences[i] = sentences[i].replace(word, "")

# stopwords'leri word_index'ten çıkartıyoruz
for word in stopwords:
    if word in word_index:
        del word_index[word]

# stopwords'leri sequences'ten çıkartıyoruz
for i in range(len(sequences)):
    for word in stopwords:
        if word in sequences[i]:
            sequences[i].remove(word)

# stopwords'leri padded'den çıkartıyoruz
for i in range(len(padded)):
    for word in stopwords:
        if word in padded[i]:
            padded[i].remove(word)

print(word_index)
print(sequences)
print(padded)
```
